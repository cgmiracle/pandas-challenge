```python
import pandas as pd
import os
import csv
```

```python
db = pd.read_csv('02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv')
db.head()
```

```python
#Displays total users
total_users = db['SN'].nunique()

#Create Data Frame
pd.DataFrame({
    "Total Users" : [total_users]
})
```

```python
#Display Purchase Analysis
unique_items = db['Item ID'].nunique()
avr_price = db['Price'].mean()
total_purchase = db['Purchase ID'].count()
total_revenue = db['Price'].sum()

#Create Data Frame
pa_df = pd.DataFrame({
    "Unique Items" : [unique_items],
    "Average Purchase Price" : [avr_price],
    "Total Purchases" : [total_purchase],
    "Total Revenue" : [total_revenue]
})

pa_df["Average Purchase Price"] = pa_df["Average Purchase Price"].map("${:,.2f}".format)
pa_df["Total Revenue"] = pa_df["Total Revenue"].map("${:,.2f}".format)

# Display data frame
pa_df[["Unique Items", "Average Purchase Price", "Total Purchases", "Total Revenue"]]
```

```python
#Display Gender Analysis
#Use .groupby to group into male and female players
gender_group = db.groupby(['Gender'])

#Calculate totals and percentages for genders
gender_total = gender_group['SN'].nunique()
gender_percent = gender_total / total_users 

#Create Data Frame
ga_df = pd.DataFrame ({
    "Totals" : gender_total, "Percentage" : gender_percent
})

ga_df['Percentage'] = ga_df['Percentage'].map("{:,.2%}".format)

ga_df[["Totals", "Percentage"]]
```

```python
#Purchase Analysis by gender
g_purchase = gender_group['Purchase ID'].count()
g_avr = gender_group['Price'].mean()
g_total = gender_group['Price'].sum()
g_avr_purch = g_total / gender_total

#Gender analysis data frame
ge_df = pd.DataFrame({
    "Total Purchases" : g_purchase,
    "Average Purchase Price" : g_avr,
    "Total Revenue" : g_total,
    "Average Purchase Per Person" : g_avr_purch
})
ge_df

#Displayed data in cleaner format
ge_df["Average Purchase Price"] = ge_df["Average Purchase Price"].map("${:,.2f}".format)
ge_df["Total Revenue"] = ge_df["Total Revenue"].map("${:,.2f}".format)
ge_df["Average Purchase Per Person"] = ge_df["Average Purchase Per Person"].map("${:,.2f}".format)

#Display data frame
ge_df[["Total Purchases", 
                    "Average Purchase Price", 
                    "Total Revenue", 
                    "Average Purchase Per Person"]].sort_values(["Total Purchases"], ascending=False)
```

```python
#Age demographics
#Create Bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

db["Age Labels"] = pd.cut(db['Age'], bins, labels = age_labels)

age_group = db.groupby(['Age Labels'])

age_total = age_group['SN'].nunique()
age_percent = age_total / total_users * 100

ad_df = pd.DataFrame({
    "Total" : age_total,
    "Percentage" : age_percent
})


ad_df["Percentage"] = ad_df["Percentage"].map("{:.2f}%".format)

#Demographics Table
ad_df[["Total", "Percentage"]]
```

```python
#Age analysis
a_purchase = age_group['Purchase ID'].count()
a_avr = age_group['Price'].mean()
a_total = age_group['Price'].sum()
a_avr_purch = a_total / age_total

#Age analysis data frame
a_df = pd.DataFrame({
    "Total Purchases" : a_purchase,
    "Average Purchase Price" : a_avr,
    "Total Revenue" : a_total,
    "Average Purchase Per Person" : a_avr_purch
})


#Clean Formatting
a_df["Average Purchase Price"] = a_df["Average Purchase Price"].map("${:,.2f}".format)
a_df["Total Revenue"] = a_df["Total Revenue"].map("${:,.2f}".format)
a_df["Average Purchase Per Person"] = a_df["Average Purchase Per Person"].map("${:,.2f}".format)

#Analysis Table
a_df[["Total Purchases", "Average Purchase Price", "Total Revenue", "Average Purchase Per Person"]]
```

```python
#CHECK TOP 5 SORT

#Top Spenders 
#Identify top 5 spenders 
top_spend = db.groupby(['SN'])

ts_purchase = top_spend['Purchase ID'].count()
ts_avr_price = top_spend['Price'].mean()
ts_total = top_spend['Price'].sum()

#Create Data Frame
ts_df = pd.DataFrame({
    "Purchase Count" : ts_purchase,
    "Average Purchase" : ts_avr_price,
    "Total Purchase Revenue" : ts_total
})

#Sort from Highest to lowest
ts_df = ts_df.sort_values(['Total Purchase Revenue'], ascending = False)

#Clean formatting
ts_df["Average Purchase"] = ts_df["Average Purchase"].map("${:,.2f}".format)
ts_df["Total Purchase Revenue"] = ts_df["Total Purchase Revenue"].map("${:,.2f}".format)

#Display data frame
ts_df[["Purchase Count", "Average Purchase", "Total Purchase Revenue"]].head()
```

```python
#Most Popular Items 
#Identify top 5 most popular items
items = db.groupby(['Item ID', "Item Name", "Price"])

items_purchase = items['Item ID'].count()
items_revenue = items['Price'].sum()

#Pop items data frame
pi_df = pd.DataFrame({
    "Purchase Count" : items_purchase,
    "Purchase Revenue" : items_revenue
})


pop_items = pi_df.sort_values(['Purchase Count'], ascending = False)

#Cleaner Formatting
pop_items["Purchase Revenue"] = pop_items["Purchase Revenue"].map("${:,.2f}".format)

#Display data frame
pop_items[["Purchase Count", "Purchase Revenue"]].head()
```

```python
#Most Profitable Items 
#Identify top 5 most profitable items
items = db.groupby(['Item ID', "Item Name", "Price"])

items_purchase_prof = items['Item ID'].count()
items_revenue_prof = items['Price'].sum()

#Pop items data frame
pi_prof_df = pd.DataFrame({
    "Purchase Count" : items_purchase_prof,
    "Purchase Revenue" : items_revenue_prof
})


prof_items = pi_prof_df.sort_values(['Purchase Revenue'], ascending = False)

#Clean formatting
prof_items["Purchase Revenue"] = prof_items["Purchase Revenue"].map("${:,.2f}".format)

# Display Data frame
prof_items[["Purchase Count", "Purchase Revenue"]].head()
```
